```python
numero1 = int(input("Ingresa el primer numero: "))
numero2 = int(input("Ingresa el segundo numero: "))
suma = numero1 + numero2

if (suma <200) :
  print("menor a 200")
elif (suma ==200):
  print("igual a 200")
elif (suma >200 and suma <=250):
  print("mayor a 200")
else:
  print("mayor a 250")
```
